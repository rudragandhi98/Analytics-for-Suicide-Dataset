```python
import numpy as np
import pandas as pd
import math
import random
import statistics as stat
```

```python
t = pd.read_csv("who_suicide_statistics.csv")




t.describe()
```

```python
t.head()
```

```python


Total_suicide_count = len(t.suicides_no)

suicide_count = [n for n in t.suicides_no if not math.isnan(n)]
valid_count = len(suicide_count)

print(f"Total suicide count is {Total_suicide_count}")

print(f"valid suicide count :{valid_count}")


```

```python
mean_suici = stat.mean(suicide_count)
stdev_or = stat.pstdev(suicide_count)
mean_suicide = round(mean_suici,2)
stdev_ori = round(stdev_or,2)
print(mean_suicide)



print(f'stat.mean      = {stat.mean(suicide_count):.2f}')
print(f'stat.median    = {stat.median(suicide_count):.2f}')
print(f'stat.pvariance = {stat.pvariance(suicide_count):.2f}')
print(f'stat.pstdev    = {stat.pstdev(suicide_count):.2f}')
```

<H3>COMPARING SAMPLE MEANS WITH THE MEAN OF THE WHOLE POPULATION USING PIECHARTS</H3>

```python
def show_pie_chart(title, wedge_labels, wedge_values):
    """
    Display a pie chart.
    @param title the chart title.
    @param wedge_labels the labels for the wedges
    @param wedge_values the values for the wedges
    """
    
    print(title)
    
    df_count_pie = pd.DataFrame({'Class' : wedge_labels, 
                                 'Counts' : wedge_values})
    df_count_pie.Counts.groupby(df_count_pie.Class).sum().plot(kind='pie')
    plt.axis('equal')
    plt.show()
```

```python
import seaborn as sns

classes = ["Original",'Sample_50', 'Sample_500', 'Sample_10000']
counts  = [mean_suicide,mean_x,mean_y,mean_z]

show_pie_chart('Sampling Mean variation ', classes, counts)

```

<b>Pie chart shows how the original and other samples except the original 500 one are pretty similarly distributed</b>
